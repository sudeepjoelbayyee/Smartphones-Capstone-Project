```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python

brands = ['samsung','xiaomi','vivo','realme','motorola','oneplus','oppo','iqoo','poco','apple','infinix','tecno','honor','nothing','nokia','itel','google','lava','huawei','nubia','asus','sony','ikall','redmagic','meizu','tesla','lg','jio','lenovo','gionee','micromax','htc','blackberry']

df = pd.DataFrame(columns=['links','model','price','rating','sim','processor','ram','battery','display','camera','card','os'])
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

df = pd.DataFrame()  # Initialize an empty DataFrame outside the loop

for brand in brands:  # Use a more Pythonic way to iterate over the brands
    with open(f'html/{brand}.html', 'r', encoding='utf-8') as f:
        html = f.read()
        soup = BeautifulSoup(html, 'html.parser')

        images,links,names, prices, ratings, sim, processor, ram, battery, display, camera, card, os = [], [], [], [], [], [], [], [], [], [], [], [], []

        for i in soup.find_all('div', {'class': 'sm-product has-tag has-features has-actions'}):
            try:
                images.append(i.find('img').get('src'))
            except AttributeError:
                images.append(np.nan)
            try:
                links.append(i.find('a').get('href'))
            except AttributeError:
                links.append(np.nan)
            try:
                names.append(i.find('h2').text)
            except AttributeError:
                names.append(np.nan)
            try:
                prices.append(i.find('span', {'class': 'price'}).text)
            except AttributeError:
                prices.append(np.nan)
            try:
                ratings.append(i.find('div', {'class': 'score rank-2-bg'}).find('b').text)
            except AttributeError:
                ratings.append(np.nan)

            x = i.find('ul', {'class': 'sm-feat specs'}).find_all('li')
            try:
                sim.append(x[0].text)
            except IndexError:
                sim.append(np.nan)
            try:
                processor.append(x[1].text)
            except IndexError:
                processor.append(np.nan)
            try:
                ram.append(x[2].text)
            except IndexError:
                ram.append(np.nan)
            try:
                battery.append(x[3].text)
            except IndexError:
                battery.append(np.nan)
            try:
                display.append(x[4].text)
            except IndexError:
                display.append(np.nan)
            try:
                camera.append(x[5].text)
            except IndexError:
                camera.append(np.nan)
            try:
                card.append(x[6].text)
            except IndexError:
                card.append(np.nan)
            try:
                os.append(x[7].text)
            except IndexError:
                os.append(np.nan)

        temp_df = pd.DataFrame({
            'images':images,
            'links':links,
            'model': names,
            'price': prices,
            'rating': ratings,
            'sim': sim,
            'processor': processor,
            'ram': ram,
            'battery': battery,
            'display': display,
            'camera': camera,
            'card': card,
            'os': os
        })
        df = pd.concat([df, temp_df], ignore_index=True)  # Concatenate the temporary DataFrame to df

# Now you have your final DataFrame df containing all the data from the HTML files
```

```python
def prepend_domain(url):
    return 'www.smartprix.com' + url

# Apply the function to the 'URL' column
df['links'] = df['links'].apply(lambda x: prepend_domain(x))
```
