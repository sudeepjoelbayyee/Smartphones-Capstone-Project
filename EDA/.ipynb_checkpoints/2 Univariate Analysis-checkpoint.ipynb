### Price

```python
df['price'].isnull().sum()
```

```python
df['price'].describe()
```

```python
sns.histplot(df['price'],kde=True,bins=50)
```

```python
df['price'].skew()
```

```python
sns.boxplot(x=df['price'],color='lightgreen')
```

```python
ecdf = df['price'].value_counts().sort_index().cumsum() / len(df['price'])
plt.plot(ecdf.index,ecdf,marker='.',linestyle='none')
```

### Observations
 - No missing values
 - Median is be correct i.e., 19990 but the mean seems to be wrong as the data is skewed due to high prices of branded phones.
 - Standard deviation seems to be also too high
 - Max value is also high
 
### Visualisations
 - The Distributions seems to be right skewed with the skewness of 2.30
 - Box plot says that there are outliers above 75000 but they are not actually considered as outliers as the price is dependent on brand.
 - With the help of CDF plot we can justify that 80% of the data is below 75000 and 20% of data is above 75000

```python
plt.figure(figsize=(15,6))
           
# Distribution without log transformation
plt.subplot(1,2,1)
sns.histplot(df['price'],kde=True,bins=50,color='skyblue')
plt.title("Distribution of Prices (Original)")

# Distribution with log transformation
plt.subplot(1,2,2)
sns.histplot(np.log1p(df['price']),kde=True,bins=50,color='lightgreen')
plt.title("Distribution of Prices (Log transformed)")
```

### Observations
 - The original Distribution of Prices are more right skewed but the log transformed distribution of Prices are more like a normal curve transforming the range from 8.0 to 12.0

```python
plt.figure(figsize=(15,6))
           
# Distribution without log transformation
plt.subplot(1,2,1)
sns.boxplot(df['price'],color='skyblue')
plt.title("Boxplot of Prices (Original)")

# Distribution with log transformation
plt.subplot(1,2,2)
sns.boxplot(np.log1p(df['price']),color='lightgreen')
plt.title("Boxplot of Prices (Log transformed)")
```

### Observations
 - The Boxplot seems to also be more improved with no outliers showing after log transformation

### Observations

 - The description of the data seems to be normal

 - But the Boxplot says it has no outliers
 - As there are no outliers and almost normally distrubuted, there is no need of transformation but checked if there will be any changes and got to know there are not changes in distribution and box plot after log transformation
